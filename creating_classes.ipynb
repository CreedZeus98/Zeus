```python
class Student:
    def __init__(self, name, college, qualification):               # dunder effect (starts with __ and ends with __. Eg - __init__)
        self.name = name
        self.college = college
        self.qualification = qualification

    def __repr__(self):                                             # __repr__ is for represent the data in list
        return self.name
```

```python
stdObjs =[ 
    Student('Upendra','MIT','Bachelors'),                       # This is called Abstraction
    Student('Zeus','IIT','Bachelors'),
    Student('Kratos','NIT','Bachelors'),
    Student('Logan','IIM','Bachelors'),
]
```

```python
class triangle:                                                 # This is called Incapsulation
    def __init__(self, b=0,h=0):
        self.base = b                                           # these are the properties of class 
        self.height = h
    
    def area(self):
        return self.base * self.height / 2
    
    def __repr__(self):                                 
        return f"traingle({self.area()})"
    
    
```

```python
data = [triangle(12,5),triangle(10,5),triangle(8,15),triangle(6,25)]
data
```

```python
import os
class reader:
    """Reads a file and returns the content"""
    def __init__(self,path):
        if not os.path.exists(path):
            raise Exception('File does not exist')
        self.path = path
        self.content = self.read()

    def read(self):
        with open(self.path,'r') as f:
            return f.read()
        
    def count_words(self):
        return len(self.content.split())
    
    def count_chars(self):
        return len(self.content)
    
    def __str__(self):
        return self.content
    
```
